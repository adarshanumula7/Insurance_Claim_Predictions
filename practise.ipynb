```python
import numpy as np

def z_score_outlier(x):
    x = np.array(x)
    mean = np.mean(x)
    std = np.std(x)
    z_scores = (x - mean) / std
    return [1 if np.abs(z) <= 3 else 0 for z in z_scores]

def IQR_outlier(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = Q3 - Q1
    Iqr_range = (Q1 - IQR * 1.5, Q3 + IQR * 1.5)
    return [1 if val >= Iqr_range[0] and val <= Iqr_range[1] else 0 for val in x]
    
```
